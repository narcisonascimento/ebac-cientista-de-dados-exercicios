````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

%matplotlib inline
````

````python
# carregando a base
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop(['Unnamed: 0', 'index'], axis=1)
df.head()
````

````python
# verificando a correlação
df.corr()
````

````python
ax = sns.pointplot(
    data=df,
    x='posse_de_veiculo',
    y='renda',
    dodge=True,
    ci=95
)
# plt.legend(loc='lower center', bbox_to_anchor=(0.5, -.30), ncol=3)
````

````python
ax = sns.pointplot(
    data=df,
    x='posse_de_imovel',
    y='renda',
    # hue='mau',
    dodge=True,
    ci=95
)
# plt.legend(loc='lower center', bbox_to_anchor=(0.5, -.30), ncol=3)
````

Insight
---
Ambas as variáveis, posse de veículo e posse de imóvel, quando verdadeiro maior é a renda, quando falso, menor é a renda.
