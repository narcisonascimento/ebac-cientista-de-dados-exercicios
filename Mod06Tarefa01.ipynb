```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
sns.countplot(x='APGAR1', data=df)
```

```python
sns.countplot(x='APGAR5', data=df)
```

```python
sns.countplot(x='SEXO', data=df)
```

```python
sns.countplot(x='GESTACAO', data=df)
```

```python
sns.countplot(x='ESTCIVMAE', data=df)
```

```python
plt.close('all')

plt.rc('figure', figsize=(15, 8))
fig, axes = plt.subplots(2, 3)

chart_1 = sns.countplot(ax = axes[0, 0], x='APGAR1', data=df)
chart_2 = sns.countplot(ax = axes[0, 1], x='APGAR5', data=df)
chart_3 = sns.countplot(ax = axes[0, 2], x='SEXO', data=df)
chart_4 = sns.countplot(ax = axes[1, 0], x='GESTACAO', data=df)
chart_5 = sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=df)

chart_4.set_xticklabels(chart_4.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.subplots_adjust(wspace=0.3, hspace=0.3)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y='CONSULTAS', data=df)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

plt.rc('figure', figsize=(15, 10))
fig, axes = plt.subplots(2, 2)

chart_1 = sns.countplot(ax = axes[0, 0], x='SEXO', data=df)
chart_2 = sns.countplot(ax = axes[1, 0], x='CONSULTAS', data=df)
chart_3 = sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=df)
chart_4 = sns.countplot(ax = axes[1, 1], x='GESTACAO', data=df)

chart_4.set_xticklabels(chart_4.get_xticklabels(), rotation=45, horizontalalignment='right')

plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
