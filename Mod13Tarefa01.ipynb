#### Previsão de renda

Vamos trabalhar com a base 'previsao_de_renda.csv', que é a base do seu próximo projeto. Vamos usar os recursos que vimos até aqui nesta base.

|variavel|descrição|
|-|-|
|data_ref                | Data de referência de coleta das variáveis |
|index                   | Código de identificação do cliente|
|sexo                    | Sexo do cliente|
|posse_de_veiculo        | Indica se o cliente possui veículo|
|posse_de_imovel         | Indica se o cliente possui imóvel|
|qtd_filhos              | Quantidade de filhos do cliente|
|tipo_renda              | Tipo de renda do cliente|
|educacao                | Grau de instrução do cliente|
|estado_civil            | Estado civil do cliente|
|tipo_residencia         | Tipo de residência do cliente (própria, alugada etc)|
|idade                   | Idade do cliente|
|tempo_emprego           | Tempo no emprego atual|
|qt_pessoas_residencia   | Quantidade de pessoas que moram na residência|
|renda                   | Renda em reais|

```python
import pandas as pd
import seaborn as sns
import numpy as np

import patsy

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# carregando e tratando os dados da base
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop(['index', 'Unnamed: 0', 'data_ref'], axis=1)
df.head()
```

```python
df.info()
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.
    

```python
reg = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + C(qtd_filhos, Treatment(0)) + C(tipo_renda, Treatment(3)) + C(educacao) + C(estado_civil, Treatment(4)) + C(tipo_residencia, Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia + mau', data=df).fit()
reg.summary()
```

2. Remova a variável menos significante e analise:
    - Observe os indicadores que vimos, e avalie se o modelo melhorou ou piorou na sua opinião.
    - Observe os parâmetros e veja se algum se alterou muito.

```python
# retidando a variável estado_civil
reg_2 = smf.ols('np.log(renda) ~ C(sexo) + posse_de_veiculo + posse_de_imovel + C(qtd_filhos, Treatment(0)) + C(tipo_renda, Treatment(3)) + C(educacao) + C(tipo_residencia, Treatment(1)) + idade + tempo_emprego + qt_pessoas_residencia + mau', data=df).fit()
reg_2.summary()
```

Análise
---
Retirando a variável estado_civil, os parâmetros não sofreram alterações.

3. Siga removendo as variáveis menos significantes, sempre que o *p-value* for menor que 5%. Compare o modelo final com o inicial. Observe os indicadores e conclua se o modelo parece melhor. 

```python
# retidando variáveis
reg_3 = smf.ols('np.log(renda) ~ C(sexo) + posse_de_imovel + idade + tempo_emprego + qt_pessoas_residencia', data=df).fit()
reg_3.summary()
```

Análise
---
O modelo 3 está mais simples, com variáveis fazendo "mais sentido", teve aumento no R², porém, o AIC está relativamente mais alto. <br/>
Entre os dois modelos, o último parece melhor mesmo com o AIC mais alto.

```python
print(f"R² Modelo 1: {reg.rsquared*100:.2f}%")
print(f"R² Ajustado Modelo 1: {reg.rsquared_adj*100:.2f}%")
print(f"AIC Modelo 1: {reg.aic:.2f}")

print("")

print(f"R² Modelo 2: {reg_2.rsquared*100:.2f}%")
print(f"R² Ajustado Modelo 2: {reg_2.rsquared_adj*100:.2f}%")
print(f"AIC Modelo 2: {reg_2.aic:.2f}")

print("")

print(f"R² Modelo 3: {reg_3.rsquared*100:.2f}%")
print(f"R² Ajustado Modelo 3: {reg_3.rsquared_adj*100:.2f}%")
print(f"AIC Modelo 3: {reg_3.aic:.2f}")
```
