````python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
# carregando a base CSV
df = pd.read_csv('previsao_de_renda.csv')
df = df.drop(['Unnamed: 0', 'index'], axis=1)
df.head()
````

````python
# verificando os tipos de dados das variáveis
df.dtypes
````

````python
# alterando data_ref (object) para datetime 
df['data_ref'] = pd.to_datetime(df['data_ref'])
df.dtypes
````

````python
# verificando a quantidade de missings
df.isnull().sum()
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# sexo
ax = sns.countplot(
    data=df,
    x='data_ref',
    hue='sexo'
)

tick_labs = df['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()

ticks = ax.set_xticks(list(range(df['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation=90)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
````
